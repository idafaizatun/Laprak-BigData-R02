## Big Data
Big Data adalah konsep pengelompokan data yang memiliki ukuran yang sangat besar dan terdiri dari data-data terstruktur, semi struktur, dan tidak terstruktur yang dapat berkembang sesuai dengan perkembangan zaman.
Big Data digunakan untuk menganilsis, memprediksi tren pasar atau mengevaluasi riisko investasi suatu perusahaan atau organisasi, sehingga dapat menghasilkan keputusan yang lebih cerdas dan strategis.

## Karakteristik Big Data

1. Volume
   Volume big data akan terus berkembang seiring berjalannya waktu, hal tersebut menyebabkan volume menjadi salah satu karakteristik dari big data.
2. Velocity (kecepatan)
   Velocity merupakan faktor penting dalam big data. Pada big data, kecepatan yang dimaksud yaitu seberapa cepat perusahaan memperoleh, menyimpan, dan mengelola data baru. Sehingga, dengan banyaknya data yang dihasilkan secara real-time.
3. Veriety
   Karakter selanjutnya yaitu variety. Karakter data ini merupakan sekumpulan big data yang bervariasi di setiap platform. Ragam data dikelompokkan menjadi tiga yaitu structured data, semi structured data, dan unstructured data. Contohnya terdapat pada media sosial yang menampilkan banyak macam data, mulai dari foto, video, formulir, hingga filter-filter unggulan. 
4. Veracity
   Varacity merupakan keakuratan data dalam memproses sebuah program. Apabila terdapat banyak data yang tidak memiliki akurasi tinggi, maka akan terjadi error. Contoh penerapannya yaitu pada nama akun Facebook, Instagram, dan lainnya. 
5. Value
   Big data dibuat agar bermanfaat bagi pengguna. Dilansir dari Hootsuite, cerita Instagram telah dipakai 500 juta pengguna nya per hari. Artinya, fitur tersebut cukup berpengaruh dan memiliki value tersendiri.

### 3. Praktik PySpark Lanjutan
- **Latihan 1**: Memanipulasi Data dengan PySpark.

````python
from pyspark.sql import SparkSession

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data, columns)

# Menampilkan DataFrame
df.show()

# Filtering data
df_filtered = df.filter(df['Usia'] > 30)
df_filtered.show()

# Menghitung rata-rata usia
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Usia")).show()

# Mengurutkan data berdasarkan usia
df_sorted = df.orderBy("Usia", ascending=False)
df_sorted.show()
````

- **Tugas 2**: Lakukan filter, penghitungan rata-rata, dan pengurutan data menggunakan PySpark.

````python
from pyspark.sql import SparkSession

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Doni", 160), ("Lia", 152), ("Gracy", 157), ("Ferry", 165)]
columns = ["Nama", "Tinggi"]
df = spark.createDataFrame(data, columns)

# Menampilkan DataFrame
df.show()

# Filtering data
df_filtered = df.filter(df['Tinggi'] > 155)
df_filtered.show()

# Menghitung rata-rata usia
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Tinggi")).show()

# Mengurutkan data berdasarkan usia
df_sorted = df.orderBy("Tinggi", ascending=False)
df_sorted.show()
````

- **Tugas 5**: Gunakan metode ini untuk menggabungkan data yang Anda buat di PySpark dengan data dari Pandas, kemudian lakukan analisis sederhana seperti menghitung rata-rata usia.

````python
import pandas as pd
from pyspark.sql import SparkSession

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data, columns)

# Membuat DataFrame Pandas
data_pandas = {"Nama": ["Ali", "Budi", "Citra", "Dina"], "Usia": [34, 23, 29, 45]}
df_pandas = pd.DataFrame(data_pandas)

# Mengonversi DataFrame dari PySpark ke Pandas
df_pandas_from_spark = df.toPandas()

# Menggabungkan DataFrame Koversi PySpark to Pandas dan Pandas
df_joined = pd.concat([df_pandas, df_pandas_from_spark], ignore_index=True)

# Mengkonversi hasil gabungan DataFrame to Pandas dan Pandas
df_spark_from_pandas = spark.createDataFrame(df_joined)
df_spark_from_pandas.show()

# Menghitung Rata-rata usia
df_spark_from_pandas.groupBy().agg(avg("Usia")).show()

````

**PENJELASAN** 

- Mengimpor pandas dengan inisial pd

- Mengimpor SparkSession dari modul pyspark.sql

- Kemduian membuat objek SparkSession untuk memulai sesi Spark.

- SparkSesiion berfungsi sebagai titik awal untuk bekerja dengan DataFrame dan SQL di Spark, menetapkan nama aplikasi Spark dengan fungsi **appName("BigDataPractice")**, dan **get)rCreat()** berfungsi untuk membuat sesi baru jika belum ada atau mendapatkan sesi jika sudah ada.
Selanjutnya membuat DataFrame sederhana yang dibuat dari data yang berisi nilai dari nama dan usia. DataFrame tersebut memiliki dua kolom yaitu **Nama** dan **Usia**

- Membuat DataFrame pandas dengan data yang sama seperti sebelumnya. 

- Lalu, mengkonversi DataFrame dari PySpark ke Pandas menggunakan metode **toPandas()**

- Menggabungkan dua DataFrame Pandas yang dikonversi dari PySpark.

- Setelah penggabungan, DataFrame Pandas yang dihasilkan dikonversi kembali menjadi DataFrame PySpark menggunakan **spark.createDataFrame(df_joined)**, dan tampikan data dengan perintah **show()**

- Terkahir yaitu menghitung rata-rata usia.

- **Tugas 6**: Gabungkan data dari PySpark dan Pandas, lalu lakukan operasi statistik seperti menghitung nilai maksimum usia.

````python
# Mengonversi DataFrame dari PySpark ke Pandas
df_pandas_from_spark = df.toPandas()

# Menggabungkan DataFrame Koversi PySpark to Pandas dan Pandas
df_joined = pd.concat([df_pandas, df_pandas_from_spark], ignore_index=True)

# Menghitung statistik deskriptif
print(df_joined.describe())

# Menghitung nilai maksimal usia
print("\n")
print(df_joined.max())
````

**PENJELASAN** 

Pertama yaitu mongonversi DataFrame dari PySpark ke Pandas

Variabel tersebut berisi **pd.concat([df_pandas, df_pandas_from_spark], ignore_index=True)**. Disini fungsi dari concat() yaitu untuk menggabungkan dua DataFrame, dan fungsi dari ignore_index=True yaitu untuk menggabungkan data dan index agar data tidak di reset dari awal. 
Setelah itu, menghitung statistik deskriptif menggunakan fungsi **describe()**.
